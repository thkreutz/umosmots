```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
# Read loss history of the model
df = pd.read_csv("models/1683187800242/checkpoint_history.txt", names=["Timestamp", "loss"], delim_whitespace=True)
```

```python
# Plot rolling mean of the loss to average over batches in each epoch
# For more epochs, one should average the loss over the full epoch.
# In our case, we only trained for 2 epochs, so we average the loss of N consecutive batches.
plt.plot(df.loss.rolling(50).mean())
```
